```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
import pandas as pd
```

```python
# Importing the Brown Corpus
from nltk.corpus import brown
```

```python
# Finding the categories
brown.categories()
```

```python
# Exploring the list of identifiers
nltk.corpus.brown.fileids()
```

```python
# Finding the tokens
from nltk import word_tokenize
tokens = word_tokenize(brown.raw())
tokens
```

```python
# Finding the size of the corpus (total number of tokens)
from nltk.probability import FreqDist
fdist = FreqDist(brown.words())
fdist.N()
```

```python
# Finding the number of tokens occurring in the corpus of each type
from nltk.probability import FreqDist
fdist = FreqDist(brown.words())
fdist
```

```python
# Tagging each word
pos_tags = nltk.pos_tag(brown.words())
pos_tags
```

```python
# Counting different types of tokens/words
from collections import Counter
counts = Counter( tag for word,  tag in pos_tags)
print(counts)
```

```python
# Finding the number of words in the "government" category
fdist["government"]
```

```python
# Finding the frequency of the tokens and printing the five most occuring tokens
freq = nltk.FreqDist(brown.words())
print("Five most frequent tokens are ", freq.most_common(5))
```

```python
# Counting the total number of sentences in the corpus
print("The number of sentences =", len(brown.sents()))
```

# 1. Reuters:
The Reuters Corpus contains 10,788 news documents totaling 1.3 million words. The documents have been classified into 90 topics, and grouped into two sets, called "training" and "test"; thus, the text with fileid 'test/14826' is a document drawn from the test set. This split is for training and testing algorithms that automatically detect the topic of a document.

```python
nltk.download('reuters')
from nltk.corpus import reuters
reuters.fileids()
reuters.categories()
```
